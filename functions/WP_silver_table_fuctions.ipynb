### Function to clean data from raw table and load it into silver table

```
@logger
def get_process_table(raw_weather_df = None):
    
    # importing required libraries
    from pyspark.sql.functions import col
    from datetime import datetime
    
    # defining the column structure for the processed/silver table
    columns = (
        "result.dt",
        "result.id",
        "result.name",
        "result.timezone",
        "result.sys.country",
        "result.coord.lat",
        "result.coord.lon",
        "result.main.temp",
        "result.main.temp_min",
        "result.main.temp_max",
        "result.main.pressure",
        "result.main.humidity",
        "result.visibility",
        "result.wind.speed",
        "result.wind.deg",
        "result.wind.gust",
        "result.weather.description",
        "result.weather.main",
        "result.clouds.all",
        "load_run_id",
        "created_on",
        "created_by"
    )
    
    # selecting the required columns for silver/processed table and renaming the columns to inprove readability
    raw_weather_df = raw_weather_df.select(*columns)
    cleaned_df = raw_weather_df.withColumn("description", col("description")[0]).withColumn("main", col("main")[0])
    
    rename_mapping = {
        "id" : "city_id",
        "name" : "city",
        "lat" : "latitude",
        "lon" : "longitude",
        "temp" : "temperature",
        "temp_min" : "minimum_temperature",
        "temp_max" : "maximum_temperature",
        "description" : "weather_description",
        "main" : "weather_condition",
        "speed" : 'wind_speed',
        'deg' : "wind_degree",
        'gust' : "wind_gust",
        'all' : "clouds_all"
    }
    for old, new in rename_mapping.items():
        cleaned_df = cleaned_df.withColumnRenamed(old, new)
        
    start = datetime.fromtimestamp(cleaned_df.selectExpr("min(dt)").first()[0])
    end = datetime.fromtimestamp(cleaned_df.selectExpr("max(dt)").first()[0])
    
    return cleaned_df, start, end
```
